```python
import os
import pandas as pd
```

```python
train_df = pd.read_excel('./time_series_375_prerpocess_en.xlsx', index_col=[0, 1])
train_df.head()
```

```python
test_df = pd.read_excel('./time_series_test_110_preprocess_en.xlsx', index_col=[0, 1])
test_df.head()
```

```python
print(train_df.shape)
print(test_df.shape)
```

```python
cols=['Lactate dehydrogenase', 'Hypersensitive c-reactive protein', '(%)lymphocyte', 'Admission time', 'Discharge time', 'outcome']

train_df = train_df[cols]
train_df.head()
```

```python
train_grouped = train_df.groupby(level=0)

train_path = './mortality_prediction/train/'
# 迭代处理每个分组
train_list = []
file_names , labels = [], []

for key, group in train_grouped:
    file_name = str(key) + '.csv'
    labels.append(group.iloc[0]['outcome'])
    file_names.append(file_name)
    group.drop(columns=['Discharge time', 'outcome'])
    group.to_csv(os.path.join(train_path, file_name))
```

```python
output_dir = './mortality_prediction/'
mode = 'train'

with open(os.path.join(output_dir, "{}_listfile.csv".format(mode)), "w") as listfile:
    listfile.write('stay,CovidLabel\n')
    for (x,y) in zip(file_names, labels):
        listfile.write('{},{:d}\n'.format(x, y))
```

```python
## 处理test
test_grouped = test_df.groupby(level=0)

test_path = './mortality_prediction/test/'
# 迭代处理每个分组

file_names , labels = [], []

for key, group in test_grouped:
    file_name = str(key) + '.csv'
    labels.append(group.iloc[0]['outcome'])
    file_names.append(file_name)
    group.drop(columns=['Discharge time', 'outcome'])
    group.to_csv(os.path.join(test_path, file_name))
```

```python
output_dir = './mortality_prediction/'
mode = 'test'

with open(os.path.join(output_dir, "{}_listfile.csv".format(mode)), "w") as listfile:
    listfile.write('stay,CovidLabel\n')
    for (x,y) in zip(file_names, labels):
        listfile.write('{},{:d}\n'.format(x, y))
```

```python
import glob

# 使用glob模块查找所有以.csv结尾的文件
train_files = glob.glob(os.path.join('./mortality_prediction/train/', '*.csv'))

print(len(train_files))
```

```python
from datetime import datetime
cnt = 0
cols = ['Lactate dehydrogenase', 'Hypersensitive c-reactive protein', '(%)lymphocyte']
for csv_file in train_files:
    df = pd.read_csv(csv_file)
    df['Hours'] = df.apply(lambda row: datetime.strptime(row['RE_DATE'], '%Y-%m-%d %H:%M:%S') - datetime.strptime(row['Admission time'], '%Y-%m-%d %H:%M:%S'), axis=1)
    df['Hours'] = df['Hours'].apply(lambda x: pd.Timedelta(x).total_seconds() / 3600)
    df.drop(columns=['PATIENT_ID', 'RE_DATE','Admission time','Discharge time','outcome'], inplace=True)
    df.dropna(how='all',thresh=2, inplace=True)
    # df.dropna(subset=cols, how='any', inplace=True)
    if df.shape[0] >= 1:
        df.to_csv(csv_file, index=False)
        cnt += 1
    else:
        print(csv_file)

print(cnt)
```

```python
# 使用glob模块查找所有以.csv结尾的文件
test_files = glob.glob(os.path.join('./mortality_prediction/test/', '*.csv'))

print(len(test_files))
```

```python
cnt = 0

for csv_file in test_files:
    df = pd.read_csv(csv_file)
    df['Hours'] = df.apply(lambda row: datetime.strptime(row['RE_DATE'], '%Y-%m-%d %H:%M:%S') - datetime.strptime(row['Admission time'], '%Y-%m-%d %H:%M:%S'), axis=1)
    df['Hours'] = df['Hours'].apply(lambda x: pd.Timedelta(x).total_seconds() / 3600)
    df.drop(columns=['PATIENT_ID', 'RE_DATE','Admission time','Discharge time','outcome'], inplace=True)
    df.dropna(how='all',thresh=2, inplace=True)
    if df.shape[0] >= 1:
        df.to_csv(csv_file, index=False)
        cnt += 1
    else:
        print(csv_file)

print(cnt)
```
